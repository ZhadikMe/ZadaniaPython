### Задание №6

Запишите загруженный датафрейм с доходами в виде таблицы incomes внутри файла sqlite в директории с блокнотом. Не сохраняйте индекс, в случае, если название таблицы уже есть в БД - перепишите эту таблицу.

```python
df_salary = pd.read_excel(
    './Employee_Salary_Dataset.ods',
)

engine = sql.create_engine('sqlite:///database.db')

df_salary.to_sql(
    name='Incomes',
    con=engine,
    if_exists='replace',
    index=False
)

pd.read_sql('Incomes', engine)
```
